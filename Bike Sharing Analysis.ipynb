## Data Preprocessing

```python
import pandas as pd # Importing the pandas library 
import numpy as np # Importing the numpy library 
import datetime # Importing the datetime module to work with date and time objects
```

We have collected the data for a span of approximately 12 hours for 3 days  (21st Dec, 22nd Dec, 23rd Dec) and stored in 3 different csv files for each day. We have loaded and concatenated the files into a single dataframe below:

```python
sat = pd.read_csv(r'Saturday_bike_data.csv') # Reading the CSV file with data retrived on Saturday
sun = pd.read_csv(r'Sunday_bike_data.csv') # Reading the CSV file with data retrived on Sunday
mon = pd.read_csv(r'Monday_bike_data.csv') # Reading the CSV file with data retrived on Monday

combined_bike_df = pd.concat([sat, sun, mon], ignore_index = True) # Combining the data from the three days into a single DataFrame.
```

Identifying the duplicate rows in the DataFrame and processing it accordingly:

```python
duplicates = data[data.duplicated()] # Identifying duplicate rows 
duplicates.sort_values(by = 'last_update') # Sorting the duplicate rows by the 'last_update' column
```

Converting the last_update column to datetime format for the ease of analysis:

```python
data['last_update'] = pd.to_datetime(data['last_update'], unit='ms') # Converting the 'last_update' column from milliseconds to datetime format

```

Generating new attributes for detailed analysis:

```python
data['date'] = data['last_update'].dt.date # Creating a new 'date' column by extracting the date from the 'last_update' column

data['hour'] = data['last_update'].dt.hour # Similarly generating 'hour' column
```

```python
data['latitude'] = data['position'].str.extract(r"'lat':\s*([\d\.\-]+)") # Generating a new 'latitute' column by extracting the lat from the 'position' column

data['longitude'] = data['position'].str.extract(r"'lng':\s*([\d\.\-]+)") # Similarly generating 'longitude' column
```
